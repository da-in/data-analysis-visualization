#### 라벨 입력하기
1. plt.xlabel()
2. plt.ylabel()

````python
import matplotlib.pyplot as plt # matplotlib의 pyplot import

x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y = [13,16,15,18,16,17,16]
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
plt.plot(x, y)
plt.show()
````

````python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y1 = [13,16,15,18,16,17,16]
y2 = [17,14,17,16,15,15,14]
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
plt.show()
````

#### legend(범례)
1. plt.legend()

````python
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y1 = [13,16,15,18,16,17,16]
y2 = [17,14,17,16,15,15,14]
plt.plot(x, y1, label='sold')
plt.plot(x, y2, label='On Shelves')
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
plt.legend(loc="upper left")
plt.show()
````

#### 제목 넣기
1. plt.title()

````python
import matplotlib.pyplot as plt
x = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y1 = [13,16,15,18,16,17,16]
y2 = [17,14,17,16,15,15,14]
plt.plot(x, y1, label='sold')
plt.plot(x, y2, label='On Shelves')
plt.xlabel('Daily')
plt.ylabel('No. of Books Sold')
plt.legend(loc="upper left")
plt.title('Le Petit Prince: Sold & left')
plt.show()
````

## 2. 점선 그래프

````python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, 'bo')
# bo = blue 'o' shape
plt.show()
````

## 3. Numpy를 이용한 plot

````python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([[1,1], [1,2], [1,3], [2,1], [2,2], [2,3], [3,1], [3,2], [3,3]])
# p=np.array([2.5, 2])

plt.plot(points[:,0], points[:,1], "ro")
# plt.plot(p[0], p[1], "bo")
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y, "ro-",linewidth=2, markersize=5)
# 라인과 마커
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y, "gs-",linewidth=1, markersize=5)
plt.show()
````

````python
x = np.linspace(0,10,20)
y1= x**2.0
y2= x**1.5
plt.plot(x,y1, "bo-", linewidth=2, markersize=12, label="First")
plt.plot(x,y2, "gs-", linewidth=2, markersize=12, label="Second")
plt.xlabel("X")
plt.xlabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.show()
````

#### 표준정규분포(Standard Normal Distribution)와 히스토그램
천개의 무작위 값을 생성하여 그래프 그리기

````python
y = [1,1,1,2,2,3,3,3,3,4,4]
plt.hist(y)
````

````python
import matplotlib.pyplot as pyplot
import numpy as np

x = np.random.standard_normal(size=1000)
plt.hist(x)
````

````python
import matplotlib.pyplot as pyplot
import numpy as np

x = np.random.randn(1000)
plt.hist(x)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# plt.style.use('default')
# plt.rcParams['figure.figsize'] = (6, 3)
# plt.rcParams['font.size'] = 12

a = np.random.randn(100000)
b = 2 * np.random.randn(100000) - 1
c = 4 * np.random.randn(100000) + 2

plt.hist(a, bins=100, density=True, alpha=0.5, histtype='step', label='(mean, stddev)=(0, 1)')
plt.hist(b, bins=100, density=True, alpha=0.75, histtype='step', label='(mean, stddev)=(-1, 2)')
plt.hist(c, bins=100, density=True, alpha=1.0, histtype='step', label='(mean, stddev)=(2, 4)')

plt.xlim(-15, 25)
plt.legend()
plt.show()
````
